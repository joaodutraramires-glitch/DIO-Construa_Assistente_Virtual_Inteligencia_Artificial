```python
from google import genai
from google.colab import userdata

api_key_colab = userdata.get("GEMINI_ASS")

client = genai.Client(
    api_key=api_key_colab
)
```

```python
print("===== MIA - Monitor Inteligente de Gastos com IA =====")

renda = float(input("Informe sua renda mensal: R$ "))

moradia = float(input("Gasto com moradia: R$ "))
alimentacao = float(input("Gasto com alimentação: R$ "))
transporte = float(input("Gasto com transporte: R$ "))
lazer = float(input("Gasto com lazer: R$ "))
streaming = float(input("Gasto com streaming: R$ "))
```

```python
perc_moradia = (moradia / renda) * 100
perc_alimentacao = (alimentacao / renda) * 100
perc_transporte = (transporte / renda) * 100
perc_lazer = (lazer / renda) * 100
perc_streaming = (streaming / renda) * 100
```

```python
total_gastos = (
    moradia +
    alimentacao +
    transporte +
    lazer +
    streaming
)

saldo = renda - total_gastos

percentual_comprometido = (
    total_gastos / renda
) * 100
```

```python
if saldo < 0:
    classificacao = "Crítica"

elif saldo < renda * 0.10:
    classificacao = "Atenção"

elif saldo < renda * 0.30:
    classificacao = "Boa"

else:
    classificacao = "Excelente"
```

```python
alertas = []

if perc_moradia > 35:
    alertas.append("Moradia acima do recomendado.")

if perc_alimentacao > 30:
    alertas.append("Alimentação acima do recomendado.")

if perc_transporte > 15:
    alertas.append("Transporte acima do recomendado.")

if perc_lazer > 10:
    alertas.append("Lazer acima do recomendado.")

if perc_streaming > 5:
    alertas.append("Streaming acima do recomendado.")
```

```python
prompt = f"""
Você é MIA (Monitor Inteligente de Gastos com IA).

Sua função é analisar os gastos do usuário e fornecer uma avaliação financeira clara.

REGRAS:

- Não invente informações.
- Não recomende investimentos.
- Seja objetiva e educativa.
- Explique os cálculos de forma simples.
- Considere os limites recomendados.

LIMITES:

Moradia: até 35%
Alimentação: até 30%
Transporte: até 15%
Lazer: até 10%
Streaming: até 5%

DADOS DO USUÁRIO

Renda: R$ {renda:.2f}

Moradia:
R$ {moradia:.2f}
({perc_moradia:.1f}%)

Alimentação:
R$ {alimentacao:.2f}
({perc_alimentacao:.1f}%)

Transporte:
R$ {transporte:.2f}
({perc_transporte:.1f}%)

Lazer:
R$ {lazer:.2f}
({perc_lazer:.1f}%)

Streaming:
R$ {streaming:.2f}
({perc_streaming:.1f}%)

Total de gastos:
R$ {total_gastos:.2f}

Saldo restante:
R$ {saldo:.2f}

Percentual comprometido da renda:
{percentual_comprometido:.1f}%

Classificação preliminar:
{classificacao}

Alertas encontrados:
{alertas if alertas else "Nenhum alerta identificado"}

Forneça:

1. Resumo financeiro.
2. Avaliação da saúde financeira.
3. Justificativa da classificação.
4. Gastos acima do recomendado.
5. Sugestões práticas para melhorar a organização financeira.
6. Conclusão final.
"""
```

```python
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

print(response.text)
```
